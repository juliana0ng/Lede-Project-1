```python
import json
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", '{:,}'.format)
```

# Starbucks

```python
data = pd.read_json('coffeeBGC.json')
df = pd.json_normalize(data['places'])
```

```python
df['nRating'] = (df.rating - 1) / 4
```

```python
df['nRatingCount'] = df.userRatingCount/1050
```

```python
df['totalScore'] = (0.6* df.nRating)+(0.4* df.nRatingCount)
```

```python
df.sort_values('totalScore', ascending=False).head(20)
```

```python
dfSB = df[df['displayName.text'].str.contains("Starbucks")]
```

```python
dfSB.totalScore.median()
```
