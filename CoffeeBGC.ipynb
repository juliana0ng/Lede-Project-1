# Top Coffee Shops in BGC according to Google Places data

```python
data = pd.read_json('coffeeBGC.json')
df = pd.json_normalize(data['places'])
df
```

## Create a "popularity score" by (1) normalizing rating and rating count

```python
df['nRating'] = (df.rating - 1) / 4
```

```python
df['nRatingCount'] = df.userRatingCount/1050
```

## (2) Multiply normalized rating and normalized rating count with a weighted ratio (0.6 for rating and 0.4 for rating count) and combine to get the final "popularity score" per coffee shop

```python
df['totalScore'] = (0.6* df.nRating)+(0.4* df.nRatingCount)
```

```python
dfFinal = df.sort_values('totalScore', ascending=False).head(19)
```

```python
dfFinal.to_csv('coffeeBGC.csv')
```
